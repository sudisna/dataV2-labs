# StackAPI

#### Import the necessary libraries here:

```python
# your code here
from datetime import datetime
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
questions = SITE.fetch('questions', fromdate=datetime(2020,4,1), todate=datetime(2020,5,1))
questions
answers = SITE.fetch('answers', fromdate=datetime(2020,4,1), todate=datetime(2020,5,1))
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions = SITE.fetch('questions', fromdate=datetime(2020,5,20), todate=datetime(2020,5,21), tagged='python',min=5, sort='votes')
questions
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
answers = SITE.fetch('answers', ids=[6784, 6473])
answers
```
