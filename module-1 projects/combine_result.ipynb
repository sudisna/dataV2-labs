```python
#import wijken_br.csv
wijken_br = pd.read_csv("wijken_buurten.csv", sep=',')
wijken_br
```

```python
# creating a geometry column 
wijken_br = geopandas.GeoDataFrame(
    wijken_br, geometry=geopandas.points_from_xy(wijken_br.lat, wijken_br.lon))
```
